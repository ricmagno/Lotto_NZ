```python
import pandas as pd
```

```python
lotto = '1c397932-922b-11ea-b048-0fd73f0d88c8 2'
# path = './Source/'
path = '~/Downloads/'
file_extension = '.xlsx' 
```

```python
file = path + lotto + file_extension
```

```python
df = pd.read_excel(file, index=True)
data_lotto = df[6:]
```

```python
data_lotto.head(2)
```

```python
data_lotto = data_lotto.rename(columns={'Lotto Powerball Winning Number Results': 'Draw',
                                        'Unnamed: 1': 'Date',
                                        'Unnamed: 2': '1',
                                        'Unnamed: 3': '2',
                                        'Unnamed: 4': '3',
                                        'Unnamed: 5': '4',
                                        'Unnamed: 6': '5',
                                        'Unnamed: 7': '6',
                                        'Unnamed: 8': 'Bonus',
                                        'Unnamed: 9': 'Bonus 2nd',
                                        'Unnamed: 10': 'Powerball'})
```

```python
data_lotto = data_lotto.sort_values(by=['Draw'])
data_lotto = data_lotto.reset_index(drop=True)
```

```python
# data_lotto[-1:]
data_lotto.tail(2)
```

```python
data_lotto[["Draw","1","2","3","4","5","6","Bonus"]] = data_lotto[["Draw","1","2","3","4","5","6","Bonus"]].astype(int)
```

```python
data_lotto['Date'] = pd.to_datetime(data_lotto['Date'], infer_datetime_format=True)
data_lotto
```
